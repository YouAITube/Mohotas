```python
from pylab import imshow, show, gray

image = mahotas.imread(path_image, as_grey=True)
image = image.astype(np.uint8)
gray()
# brightness threshold calculation
otsu = mahotas.otsu(image)
# output only those pixels that are brighter than the obtained threshold
imshow(image > otsu)
show()
```

```python
# increase brightness by 2 times
image_bright = mahotas.stretch(image ** 2.0)

# decrease brightness by 2 times
image_bright_1 = mahotas.stretch(image ** 0.5)
```

```python
# Assuming image_bright and image_bright_1 are numpy arrays representing images
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

axes[0].imshow(image_bright, cmap='gray')
axes[0].set_title('Increase brightness by 2 times')
axes[0].axis('off')

axes[1].imshow(image_bright_1, cmap='gray')
axes[1].set_title('Decrease brightness by 2 times')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
path_image = r"./dog_2.png"
image = mahotas.imread(path_image)

# apply a filter to the image
image = image[:, :, 0]
# blur the image
image = mahotas.gaussian_filter(image, 1)

# set threshold value
mean_val = image.mean()
image = (image > mean_val)
```

```python
# Assuming im_rgb is a numpy array representing a grayscale image
plt.imshow(image, cmap='gray')
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
# labeling areas in the image
lable, objects = mahotas.label(image)
imshow(lable)
show()

# get the size of each area
sizes = mahotas.labeled.labeled_size(lable)
```

```python
# select areas to remove
objects_to_remove = np.where(sizes < 1500)

# remove objects
image_without_objects = mahotas.labeled.remove_regions(lable, objects_to_remove)
imshow(image_without_objects)
show()
```

```python
# label the borders of the objects
border_image = mahotas.labeled.borders(image_without_objects)
imshow(border_image)
show()
```
